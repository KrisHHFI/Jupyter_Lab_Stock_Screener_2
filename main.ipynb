```python
from utils.const.filters import SCREEN_SIZE_FILTER
from utils.yfinance.call_tracker.get_yfinance_call_count import get_yfinance_call_count
from utils.yfinance.call_tracker.get_yfinance_call_log import get_yfinance_call_log
from utils.yfinance.call_tracker.reset_yfinance_call_count import reset_yfinance_call_count
from utils.yfinance.call_tracker.set_yfinance_call_delay import set_yfinance_call_delay
from utils.yfinance.get_most_active_us_stocks import get_most_active_us_stocks
from utils.printing.print_stock_info import print_stock_info
from utils.printing.print_data_collected import print_data_collected

delay_seconds = 1.5
top_n = SCREEN_SIZE_FILTER
analyst_rating_filter = "1.3 - Strong Buy"
include_next_tier_down = True

try:
    reset_yfinance_call_count()
    set_yfinance_call_delay(delay_seconds)

    stock_info = get_most_active_us_stocks(
        top_n=top_n,
        screen_size=SCREEN_SIZE_FILTER,
        verify_with_info=False,
        analyst_rating_filter=analyst_rating_filter,
        include_next_tier_down=include_next_tier_down,
    )

    print_stock_info(stock_info)
    print_data_collected(
        yfinance_call_count=get_yfinance_call_count(),
        yfinance_call_delay=delay_seconds,
        yfinance_call_log=get_yfinance_call_log(),
    )
except Exception as e:
    print(f"Error: {e}")
```
